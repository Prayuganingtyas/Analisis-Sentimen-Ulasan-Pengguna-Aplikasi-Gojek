```python
app_package = 'com.gojek.app'
all_reviews = []
target_count = 12000
batch_size = 200
cursor = None

while len(all_reviews) < target_count:
    for score in range(1, 6):
        try:
            r, cursor = reviews(
                app_package,
                lang='id',
                country='id',
                sort=Sort.NEWEST,
                count=batch_size,
                filter_score_with=score
            )
            for review in r:
                all_reviews.append({
                    'userName': review['userName'],
                    'content': review['content'],
                    'score': review['score'],
                    'at': review['at']
                })
            time.sleep(1)
        except Exception as e:
            print(f"Gagal mengambil data skor {score}: {e}")
            continue

    print(f"Total terkumpul: {len(all_reviews)} review")
```

```python
scraped_df = pd.DataFrame(all_reviews)
print("Kolom yang tersedia:", scraped_df.columns)

print(f"Jumlah ulasan yang berhasil diambil: {len(scraped_df)}")

scraped_df.head()
```
